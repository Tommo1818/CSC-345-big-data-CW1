Task 2.1:

The iris data is taken from the Fisher Iris Dataset. It contains sepal and petal length, width, and species of 150 flowers.

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster as skcl
import sklearn.mixture as skmi

data = np.load('./Iris_data.npy')
labels = np.load('./Iris_labels.npy')

# Choose two features dimensions to plot, keeping it consistent for the notebook
x_feature = 0
y_feature = 1
```

```python
plt.scatter(data[:,x_feature], data[:,y_feature], c=labels)
plt.title('Iris Data')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

Task 2.2:

The following code uses the kmeans method from the sklearn.cluster module to fit and predict the given number of clusters of the Fisher Iris Dataset, work out each cluster centre, and then show it on a scatter plot.

```python
predicted_labels = skcl.KMeans(n_clusters=3).fit_predict(data)
plt.scatter(data[:,x_feature], data[:,y_feature], c=predicted_labels)
plt.title('Iris Data With KMeans Clustering')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
# add the cluster centers to the plot
plt.scatter(skcl.KMeans(n_clusters=3).fit(data).cluster_centers_[:,x_feature], skcl.KMeans(n_clusters=3).fit(data).cluster_centers_[:,y_feature], c='red')
plt.show()
```

Task 2.3:

The following code uses the GMM clustering method from the sklearn.mixture module to fit and predict the clustering of the Fisher Iris Dataset, work out the gaussian means centres, and the show it in a plot.

```python
dataGMM = skmi.GaussianMixture(n_components=3)
predicted_labelsGMM = dataGMM.fit_predict(data)
plt.scatter(data[:,x_feature], data[:,y_feature], c=predicted_labelsGMM)
plt.title('Iris Data With GMM Clustering')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
# add the g gaussian means to the plot
plt.scatter(dataGMM.fit(data).means_[:,x_feature], dataGMM.fit(data).means_[:,y_feature], c='red')
plt.show()
```

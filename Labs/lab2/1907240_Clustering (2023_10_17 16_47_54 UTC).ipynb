Task 2.1:

The iris data is taken from the Fisher Iris Dataset. It contains sepal and petal length, width, and species of 150 flowers.

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster as skcl
import sklearn.mixture as skmi

data = np.load('./Iris_data.npy')
labels = np.load('./Iris_labels.npy')

# Choose two features dimensions to plot, keeping it consistent for the notebook
x_feature = 0
y_feature = 1
```

```python
plt.scatter(data[:,x_feature], data[:,y_feature], c=labels)
plt.title('Iris Data')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

Task 2.2:

The following code uses the kmeans method from the sklearn.cluster module to fit and predict the given number of clusters of the Fisher Iris Dataset, and then shows it on a scatter plot.

```python
predicted_labels = skcl.KMeans(n_clusters=3).fit_predict(data)
plt.scatter(data[:,x_feature], data[:,y_feature], c=predicted_labels)
plt.title('Iris Data')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```
